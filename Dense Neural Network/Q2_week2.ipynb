```python
Goalkeeper= ["GK"]
Forward_Player=["ST", "LW", "RW", "LF", "RF", "RS","LS", "CF"]
Midfielder_Player=["CM", "RCM", "LCM", "CDM", "RDM", "LDM", "CAM", "LAM",
"RAM", "RM", "LM"]
Defender_Player=["CB", "RCB", "LCB", "LWB", "RWB", "LB", "RB"]

def replace(x:str):
    x=x.upper()
    if x in Goalkeeper:
        return "Goalkeeper"
    elif x in Forward_Player:
        return "Forward"
    elif x in Midfielder_Player:
        return "MidFielder"
    elif x in Defender_Player:
        return "Defender "
```

```python
df['Position'] = df['Position'].apply(replace)
```

```python
df['Position'].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oe = OneHotEncoder()
y=oe.fit_transform(df[['Position']])
y=pd.DataFrame(y.toarray(),columns=oe.get_feature_names_out(['Position']))
```
